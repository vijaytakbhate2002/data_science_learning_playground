```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## What is PCA and Where to use it?
- **PCA (Principal Component Analysis)** is a mathematical technique of adjusting the data dimensions in a way that some of the first dimension or first principal componenet (PC1) will hold large infomration or Eigen Value of the entire dataset as compared to last dimension or last principal component (PCn).

- Where to use?
    - Feature reduction - If we are dealing with the very high dimensional dataset we can reduce the features by utilizing some of the top principal components holding high Eigen Values.
    - Identifying Clusters 
    - Representation of high dimenstional data 

```python
data = pd.read_csv("Sythetic_Correlated_Data.csv")
data.head()
```

## Let's visualize 3 dimensions of data

```python
# plot 3d scatter plot using go library
import plotly.graph_objects as go

fig = go.Figure(data=[go.Scatter3d(
    x=data['Feature_1'],
    y=data['Feature_2'],
    z=data['Feature_3'],
    mode='markers',
    marker=dict(size=5, color='blue')
)])
fig.update_layout(scene=dict(
    xaxis_title='Feature 1',
    yaxis_title='Feature 2',
    zaxis_title='Feature 3'
))
```

## Generate PCA model 
- 2 principal components (2 dimensions)
- get eigenvalues ((sum of squared distance) / (n - 1))
- get eigenvectors 

```python
from sklearn.decomposition import PCA

# generate PCA model with 2 principal components
pca_model = PCA(n_components=2)
pca_result = pca_model.fit_transform(data)

# eigenvalues and eigenvectors of each features
eigenvalues = pca_model.explained_variance_
eigenvectors = pca_model.components_

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

## Compute information retained using eigenvalues

```python
# % information retained by each principal component using eigenvalues
information_retained = [eigenvalues[i] / sum(eigenvalues) for i in range(len(eigenvalues))]
information_retained
```

## Show scree plot with % information held by each feature

```python
# Scree plot with bar's to show the information retained by each principal component using go library
fig = go.Figure(data=[go.Bar(
    x=['Principal Component 1', 'Principal Component 2'],
    y=information_retained,
    marker_color=['blue', 'orange']
)])
fig.update_layout(title='Scree Plot', xaxis_title='Principal Components', yaxis_title='Information Retained')
fig.show()
```

## Scatterplot with PC1 and PC2

```python
# plot 2d scatter plot of pca result using plotly library
fig = go.Figure(data=[go.Scatter(
    x=pca_result[:, 0],
    y=pca_result[:, 1],
    mode='markers',
    marker=dict(size=5, color='blue')
)])
fig.update_layout(title='PCA Result', xaxis_title='Principal Component 1', yaxis_title='Principal Component 2')
fig.show()
```
