```python
import pandas as pd
import numpy as np
```

```python
# Parameters
n = 10        # number of trials
p = 0.5       # probability of success
size = 1000   # number of samples

# Generate binomial data
data = np.random.binomial(n=n, p=p, size=size)

# Convert to DataFrame
df = pd.DataFrame({'successes': data})

print(df.head())
```

```python
df['successes'].value_counts().sort_index()
```

```python
import matplotlib.pyplot as plt

plt.hist(data, bins=range(n+2), edgecolor='black')
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.title("Binomial Distribution")
plt.show()
```

```python
from scipy.stats import binom

# Probability of exactly k successes
k = 6
prob = binom.pmf(k, n, p)

print(f"P(X = {k}) =", prob)
```

```python
# Empirical probability
empirical_prob = (data == k).mean()

print("Empirical Probability:", empirical_prob)
print("Theoretical Probability:", prob)
```
