```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# 1. Load the large CSV
file_path = 'home-credit-default-risk\\bureau.csv'
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
first_1000= df.head(1000)
first_1000.rename(columns={col: col.lower().strip() for col in first_1000.columns}, inplace=True)
first_1000.head()
```

```python
applications = pd.read_csv('home-credit-default-risk\\application_train.csv')
apps = applications[applications['SK_ID_CURR'].isin(first_1000['sk_id_curr'])]
apps.head()
```

```python
apps.shape
```

```python

# 2. Create connection string
# Format: postgresql://username:password@localhost:5432/practice_db
engine = create_engine('postgresql://postgres@localhost:5432/practice_db')
```

```python
# 3. Upload and create table automatically
# 'if_exists=replace' will create the table or overwrite it
apps.to_sql('applications', engine, index=False, if_exists='replace', chunksize=10000)

print("Upload complete!")
```
