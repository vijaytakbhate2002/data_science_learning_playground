```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# 1. Load the large CSV
file_path = 'home-credit-default-risk\\bureau.csv'
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python

# 2. Create connection string
# Format: postgresql://username:password@localhost:5432/practice_db
engine = create_engine('postgresql://postgres@localhost:5432/practice_db')
```

```python

# 3. Upload and create table automatically
# 'if_exists=replace' will create the table or overwrite it
df.to_sql('my_large_table', engine, index=False, if_exists='replace', chunksize=10000)

print("Upload complete!")
```
