# Types of RAG (Retrieval-Augmented Generation)

## 1. Naive RAG
Naive RAG is the simplest form of retrieval-augmented generation where documents are embedded into a vector database, and the most similar chunks are retrieved based on a query. These retrieved chunks are then directly passed to the language model to generate a response. It does not include any optimization, re-ranking, or reasoning mechanisms.

---

## 2. Advanced RAG (Optimized Retrieval)
Advanced RAG improves retrieval quality by incorporating techniques such as hybrid search (BM25 [Best match] + embeddings), query rewriting, re-ranking, and better chunking strategies. It aims to retrieve more relevant and precise context before passing it to the language model, resulting in more accurate outputs.
> Query → (BM25 + Embeddings in parallel) → Merge → Top-K → LLM

---

## 3. Modular RAG
Modular RAG breaks the RAG pipeline into independent components such as retriever, ranker, and generator. Each module can be independently developed, replaced, or optimized, making the system flexible, scalable, and suitable for production environments.

---

## 4. Graph RAG
Graph RAG uses knowledge graphs to represent data as entities and relationships instead of flat text chunks. Retrieval is performed through graph traversal, enabling multi-hop reasoning and better handling of complex queries involving relationships between entities.

---

## 5. Agentic RAG
Agentic RAG introduces an agent (decision-making layer) that dynamically decides when to retrieve information, what to retrieve, and how many steps to take. It enables multi-step reasoning, tool usage, and iterative retrieval for complex problem-solving tasks.

---

## 6. Multimodal RAG
Multimodal RAG extends traditional RAG by handling multiple data types such as text, images, tables, and PDFs. It retrieves and processes information across different modalities to generate more comprehensive and context-aware responses.

---

## 7. Corrective RAG
Corrective RAG includes a self-evaluation mechanism where the model checks the quality of retrieved information and its generated response. If the context is insufficient or incorrect, it can re-retrieve or refine the answer, reducing hallucinations and improving reliability.
