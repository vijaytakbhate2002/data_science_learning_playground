## Embedding generation step by step guide
1. Tokenization: Split sentence into words and convert each word into respective token ID (Here token ID is fixed for each token) but models like BERT subword tokenization (WordPiece / Byte Pair Encoding), So IDs map to subword units, not full words. **eg. "unbelievable" → ["un", "##believ", "##able"] -> [1012,3056,2003]** - Dim = single tokenID vector (1, Tn)
2. Convert tokens into vectors (Embedding Matrix): Every model has their own vocabulary collection (trainable paramerter matrix that maps token ID's into dense vectors). So we retrieve embeddings of each token ID using lookup operation. - Dim = (tokenID, token embedding dimensions), here embedding dimensions depends on model, for BERT 768 dimensional token embeddings.
3. Contextual transformation (Transformer magic): Models like BERT or sentence transformers update each token ID vector using self-attention - Dim = (tokenID, updated token embedding dimensions)
4. Sentence embedding (aggregation): Uses mean pooling which reduce the tokenID dimension to 1, which means we choose only max embedding value from every updated token embedding dimensions. - Dim = (1, updated token embedding dimensions)
